# EigenVectors
- Only symmetric metrices would behave the way we saw in part 2, where the semimajor and semiminor axis vectors only got scaled



```python
A = np.array([ [2,1], [1,2]])
desc_array( A)
plot_linear_transformation(A)
```

```python
## Using unit vectors around a circle or radius 1
alpha = np.linspace(0, 2*np.pi, 41)
vectors = list(zip(np.cos(alpha), np.sin(alpha) ) )
vecs_trans = [ A.dot(x) for x in vectors]
plot_vector(vectors)
plot_vector(vecs_trans)
```

```python
semi_major = max( linalg.norm(vecs_trans, axis=1))
semi_minor = min( linalg.norm(vecs_trans, axis=1))

iu = semi_major / np.sqrt(2) 
ju = semi_minor / np.sqrt(2) 

print(f"Semi-major = {semi_major} and Semi-minor = {semi_minor}")

u1 = np.array(  [ iu, iu] )
u2 = np.array( [ -ju, ju] )
v1 = linalg.inv(A).dot( u1 )
v2 = linalg.inv(A).dot( u2 )
print( f"v1 = {v1}")
print( f"v2 = {v2}")
```

# Generalizing Eigendecomposition

- In the above case $A$ is symmetric and square 

**Orthogonal Matrix**
$$ R^T = R^{-1} $$
- A matrix whose transpose is equal to its inverse

- In above example the rotation matrix is special like that. We did R.T as a way to undo the 45 degree rotation instead of R.inv b/c the two are the same


- The columns of an orthogonal matrix are orthogonal vectors of unit length i.e. **orthonormal**
    - Unit Length: 

- Eigenvectros of a symmetric matrix are orthogonal. In example above, eigenvectors of $A$ are orthogonal


**General Case** when not symmetric or 
- With $A$ the eigenvectors land on their span; they only get scaled 
$$ A\mathbf{v}_1 = s_1 \mathbf{v}_1 $$
$$ A\mathbf{v}_2 = s_2 \mathbf{v}_2 $$

- The RHS is a vector result from matrix-vector multiplication on the LHS. And so by principle $M.x = x_1*mcol_1 + x_2* mcol_2 ... + x_n * mcol_n$ then

$$ A [\mathbf{v}_1 \mathbf{v}_2] = [\mathbf{v}_1 \mathbf{v}_2]  [[ s_1 0][0 s_2] $$

Substituting $C$ for the eigenvectors matrix, and $D$ for the scaling diagnonal matrix (with the eigenvalues)
$$ A C = C D$$

$$ A = C D C^{-1} $$


**Diagonalizable** if a matrix $A$ can be decomposed in the above way and so $C^{-1} A C = D$

- In **Change of basis** $A$ has the same effect as $D$ and so $A$ and $D$ are similar 
- Similar matrices have the same effect, via a change of basis << via???
- then substiting $A$ with the ?eigendecomposition?
$$A\mathbf{x} = C D C^{-1} \mathbf{x} $$

## Using Py to compute the Eigens

`np.linalg.eig`
The output is
- at 0: eigen values in 1d
- at 1: eigen vectors in nD as per originam matrix A

```python
linalg.eig(A)
```

```python
printHeadline('Eigens', "Results")
print("starting with major then minor??")

evalz, evecz = linalg.eig(A)
for z in zip(evalz, evecz):
    print( f"eigen val ={z[0]}  and vec = {z[1]}" )
```

```python
printHeadline('Decomposition', "@A Results << should match A")
#A_decomposition  = C @ D @ C.inv
A_decomp = evecz @ np.diag( evalz ) @ linalg.inv(evecz) 
print("A = \n", A)
print("A-decomp = \n", A_decomp)
```

```python
printHeadline('Decomposition', "@B Results << should match A")

evalz, evecz = linalg.eig(B)
for z in zip(evalz, evecz):
    print( f"eigen val ={z[0]}  and vec = {z[1]}" )
    
#A_decomposition  = C @ D @ C.inv
B_decomp = evecz @ np.diag( evalz ) @ linalg.inv(evecz) 

print("A = \n", B)
print("A-decomp = \n", B_decomp)
```
