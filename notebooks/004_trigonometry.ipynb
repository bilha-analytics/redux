Trigonometry: relationships between side lengths and angles of a triangle

**1. Ratios**
![](https://upload.wikimedia.org/wikipedia/commons/thumb/4/4f/TrigonometryTriangle.svg/245px-TrigonometryTriangle.svg.png)
- **SOH:** $ sin(theta) = opp / hyp $
- **CAH:** $ cos(theta) = adj / hyp $
- **TOA:** $ tan(theta) = opp / adj = sin(theta) / cos(theta)$

- **Norm** computes hypotenuse length by pythagorous 

**2. Using reciprocals**
- Cosecant csc --> sin $ = 1 / sin(theta) = hyp / opp $
- Secant sec --> cos $ = 1 / cos(theta) = hyp / adj $
- Cotangent cot --> tan $ = 1 / tan(theta) = adj / opp = cos(theta) / sin(theta) $

**3. The Unit Circle**
![](https://upload.wikimedia.org/wikipedia/commons/thumb/b/b5/Sin-cos-defn-I.png/220px-Sin-cos-defn-I.png) 
- Unit circle has radius 1 and is centered at the origin. 
- So length of hypotenuse is one $hyp = 1$
- So length $adj = hyp * cos(theta) = cos( theta)$
- So length $opp = hyp * sin(theta) = sin( theta)$

**4.1. Law of Sines**
$$ a/sin(A) = b / sin(B) = c / sin(C) = 2R = abc/2D$$

**4.2. Law of Cosines**
$$ c^2 = a^2 + b^2 - 2ab cos(C) $$
$$ cos(C) = (a^2 + b^2  - c^2) / 2ab $$
- An extension of pythagorean theorem to arbitrary triangles 
- law of tangents simplifies it more for tables 

**4.3. Area**
- Given two sides a and b and the angle between them C, then the area of the triangle is
$$ D = 0.5 * a * b *sin(C) = abc/4R $$

**4.4. Pythagorean identities**
$$ sin^2(A) + cos^2(A) = 1 $$
$$ tan^2(A) + 1 = sec^2(A) $$
$$ cot^2(A) + 1 = csc^2(A) $$

```python
import threading
import time 
import random
%matplotlib notebook 
def draw_random():
    
    plt.ion() ## turns on interactive mode 
    fig = plt.figure()
    ax = fig.add_subplot(111)  
    
#     plt.show()
    
    def update_data(ax=ax, fig=fig):
        t = np.linspace(0, 16*np.pi, 1000) 
        #for _ in range(20):
        while True:
            rz = np.random.rand(2)
            r1 = rz[0]
            r2 = rz[1]
            x = np.cos(r1 * t )
            y = np.sin(r2 * t )            
            ##close loop
            x = np.append(x, x[0] )
            y = np.append(y, y[0] )
            
            c = random.choice(['k', 'b', 'r', 'g', 'y', 'o'])
            ax.clear()            
            ax.plot(x,y, c)
            fig.canvas.draw()
            
            plt.axis('square')
            plt.axis( [-1.1, 1.1, -1.1, 1.1] )
            plt.axis('off')
            plt.title( f"cos r$_1$={np.round(r1,2)}, sin r$_2$={np.round(r2, 2)}" )
            plt.show()
            
            time.sleep(3)
        
    t = threading.Thread(target=update_data, daemon=True)
    t.start()
    
   
```

```python
draw_random()
```
